```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
x = np.linspace(0,5,11)
y = x ** 2
```

```python
x
```

```python
y
```

```python
# Functional Approach
plt.plot(x,y)
```

```python
plt.plot(x,y,'r-')
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('Title')
```

```python
# Subplots
```

```python
plt.subplot(1,2,1) 
# First Argument denotes the rows => 1 means 1 row 
# Second Argumet denotes the columns =>  2 means 2 columns and 
# Third Argument denotes the plot number => 3rd argument 1 means firsr plot
plt.plot(x,y,'r')        # r means Red Color

plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

```python
# Object Oriented Approach
```

```python
fig = plt.figure()
axes = fig.add_axes([0.1,0.1,0.8,0.8])
axes.plot(x,y)
axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title('Title')
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.5,0.4,0.3]) 
# First Argument => it denotes gap from left
# Second Argument => it denotes gap from bottom
# Third Argument => Denotes Width
# Fourth Argument => Denotes Height
# In Our Case (axes2) => 1st argument means 20 prcnt from left , 50 prcnt from bottom 40 prnct of width and 30 prcnt of Height
#axes2 = fig.add_axes([0.4,0.15,0.4,0.3])

axes1.plot(x,y,'r')
axes1.set_title('LARGER PLOT')
axes2.plot(y,x,'g')
axes2.set_title('SMALLER PLOT')
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes1.plot(x,y)
```
