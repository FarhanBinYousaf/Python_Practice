```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
x = np.linspace(0,5,11)
y = x ** 2
```

```python
fig = plt.figure(figsize=(8,2))        # figsize(width,height)

ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
```

```python
# Same working as above using subplots
fig,axes = plt.subplots(figsize=(8,2))        # figsize(width,height)
axes.plot(x,y)
```

```python
fig.savefig('my_picture.jpg',dpi=200)   # This will save your fig in the form of picture in your desire format like jpg, png etc
# dpi = dor per inch => enhance the quality of picture
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
ax.set_title('Title')
ax.set_ylabel('Y')
ax.set_xlabel('X')
```

```python
fig.savefig('New_picture.png',dpi=200)
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,x**2,label='X Square')
ax.plot(x,x**3,label='X Cube')
ax.legend(loc=0)       # legend(location) it takes location as perameter, 0 is the best location,you can change it to 1,2,3,4 etc
                        # loc=10 takes the legend into center of fig
```
